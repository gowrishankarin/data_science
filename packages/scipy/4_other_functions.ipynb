```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal as mvn 
```

```python
cov = np.array([[1, 0.8], [0.8, 3]])
mu = np.array([0,2])
r = mvn.rvs(mean=mu, cov=cov, size=10000)
plt.scatter(r[:,0], r[:,1])
plt.axis('equal')
plt.show()
```

```python
# scipy.io.loadmat
# scipu.io.wavfile.read - SOUND
# scipy.signal.convolve
# FILTERS: 
# FFT in numpy -- Time to Frequency Domain
x = np.linspace(0, 100, 10000)
y = np.sin(x) + np.sin(3*x) + np.sin(5*5)
plt.plot(y)
plt.show()
```

```python
Y = np.fft.fft(y)
```

```python
plt.plot(np.abs(Y))
plt.show()
```
