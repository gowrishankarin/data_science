```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal as mvn 
```

```python
cov = np.array([[1, 0.8], [0.8, 3]])
```

```python
mu = np.array([0,2])
```

```python
r = mvn.rvs(mean=mu, cov=cov, size=10000)
plt.scatter(r[:,0], r[:,1])
plt.axis('equal')
plt.show()
```

```python
r = np.random.multivariate_normal(mean=mu, cov=cov, size=1000)
plt.scatter(r[:,0], r[:,1])
plt.axis('equal')
plt.show()
```
