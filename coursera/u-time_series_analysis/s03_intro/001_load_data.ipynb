```python
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set

DATA_DIR = '../data'
```

## Load Data

```python
raw_csv_data = pd.read_csv(os.path.join(DATA_DIR, "Index2018.csv"))
```

```python
df_comp  = raw_csv_data.copy()
df_comp.head()
```

```python
df_comp.describe()
```

```python
nas = df_comp.isna()
```

```python
nas.sum()
```

## Visualize Data

```python
df_comp.spx.plot(figsize=(20, 5), title="S&P500 Prices")
plt.show()
```

```python
df_comp.dax.plot(figsize=(20, 5), title="DAX Prices")
plt.show()
```

```python
df_comp.ftse.plot(figsize=(20, 5), title="FTSE Prices")
plt.show()
```

```python
df_comp.nikkei.plot(figsize=(20, 5), title="Nikkei Prices")
plt.show()
```

```python
df_comp.spx.plot(figsize=(20, 5), title="S&P500 Prices")
df_comp.ftse.plot(figsize=(20, 5), title="FTSE Prices")
plt.title("S&P500 vs FTSE")
plt.show()
```

## QQ Plot - Quantile Quantile Plot

```python
import scipy.stats
import pylab
```

- `Theoretical Quantiles`: How many standard deviations away from the mean these values are
- This shows the data is not normally distributed

```python
scipy.stats.probplot(df_comp.spx, plot=pylab)
pylab.show()
```
