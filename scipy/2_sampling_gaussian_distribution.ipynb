```python
%matplotlib inline
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.random.randn(10000)
```

```python
plt.hist(r, bins=100)
plt.show()
```

```python
r = 10*np.random.randn(10000) + 5
```

```python
plt.hist(r, bins=100)
plt.show()
```
