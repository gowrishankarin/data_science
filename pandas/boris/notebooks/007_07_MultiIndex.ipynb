### The .pivot Method

```python
sales = pd.read_csv("../data/salesmen.csv", parse_dates=["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
sales.head()
```

```python
sales["Salesman"].value_counts()
```

```python
new_sales = sales.pivot(index="Date", columns="Salesman", values="Revenue")
```

### The pivot_table() Method

```python
food = pd.read_csv("../data/foods.csv")
food.head()
```

```python
food.pivot_table(values="Spend", index="Gender", aggfunc="mean")
```

```python
food.pivot_table(values="Spend", index=["Gender", "Item"], columns=["Frequency", "City"], aggfunc="mean")
```

```python
food.pivot_table(values="Spend", index=["Gender", "Item"], columns="City", aggfunc=["max", "mean"])
```

```python
pd.pivot_table(data=food, values="Spend", index=["Gender", "Item"], columns="City", aggfunc="mean")
```
